<font size = 5.5> **Plot a Scatter Plot between "total_bill" and "tip"**

```python
# Scatter Plot 
sns.relplot(x = "total_bill", y = "tip", data = df, kind = "scatter")
```

```python
# Scatter Plot 
### difference - kind -- "scatter" and "line"
sns.relplot(x = "total_bill", y = "tip", data = df, kind = "line")
```

<font size = 5.5> **The "hue" parameter** -- Generally we pass a categorical data

```python
# Hue Parameter

sns.relplot(x = "total_bill", y = "tip", hue = "smoker", data = df)
```

<font size = 5.5> **The "style" parameter** -- -- Generally we pass a categorical data

```python
# Style Parameter -- without "hue"
sns.relplot(x = "total_bill", y = "tip",style = "sex", data = df)

## style = "sex"  -- showing "male" and "female" -- by different symbols.
```

```python
# Style Parameter - with "hue"
sns.relplot(x = "total_bill", y = "tip", hue = "smoker",style = "sex", data = df)
```

<font size = 5.5> **The "size" parameter**

```python
sns.relplot(x = "total_bill", y = "tip", hue = "smoker",style = "sex", 
            size = "size", data = df)
```

```python
## size - optional parmeter -- sizes -(10,100) -- by passsing a tuple 
### Use -- Discrete or continuous numerical data. 

sns.relplot(x = "total_bill", y = "tip", hue = "smoker",style = "sex", 
            size = "size", sizes = (10,100), data = df)
```

<font size = 5.5> **Using  the scatter plot function** 

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "smoker",style = "sex", 
            size = "size", data = df)
```
