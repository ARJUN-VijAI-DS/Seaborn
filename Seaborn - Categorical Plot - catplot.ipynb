```python
df = sns.load_dataset("tips")
df.head()
```

## 1. Categorical Scatterplots
- a. Strip Plot
- b. Swarm Plot

### Categorical Data on -- (x - axis)
### Numerical Data on -- (y - axis)

## Strip Plot

```python
sns.catplot(x="day", y="tip", kind ="strip", data = df)

## kind ="strip"  -- Default Value.
```

```python
sns.catplot(x="day", y="tip", kind ="strip", jitter=.1, data = df)

## kind ="strip"  -- Default Value.
## jitter -- "Noise" -- between two or more similar points
### jitter = 0.1
```

```python
sns.catplot(x="day", y="tip", kind ="strip", jitter=.2, data = df)

## kind ="strip"  -- Default Value.
## jitter -- "Noise" -- between two or more similar points
### jitter = 0.2
```

```python
sns.catplot(x="day", y="tip", kind ="strip", jitter=0, data = df)

## kind ="strip"  -- Default Value.
## jitter -- "Noise" -- between two or more similar points
### jitter = 0
```

## Swarm Plot

```python
sns.catplot(x="day", y="tip", kind ="swarm", data = df)

## kind ="strip"  -- Default Value.
## Distribution of Data
```

```python
sns.catplot(x="day", y="tip", kind ="swarm",hue = "sex",  data = df)

## kind ="strip"  -- Default Value.
## Distribution of Data
## Use of "hue" -- "Hyperparameter"
```

```python
sns.swarmplot(x="day", y="tip", hue="sex", data = df)

## Distribution of Data
## Use of "hue" -- "Hyperparameter"
```

```python
sns.swarmplot(x="day", y="tip", data = df)

## Distribution of Data
```
