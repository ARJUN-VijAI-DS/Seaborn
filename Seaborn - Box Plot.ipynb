```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset("tips")
df.head()
```

<font size = 5.5> **Boxplots** -- A box plot is a standardized way of displaying
    <font size = 4.5> 
    <br><br> 1. Minimum  -- [Q1 - 1.5*IQR] -- IQR = Q3 - Q1
    <br><br> 2. Maximum  -- [Q3 + 1.5*IQR] -- IQR = Q3 - Q1
    <br><br> 3. Q1 -- 25th percentile
    <br><br> 4. Q2 -- "Median"  -- 50th percentile
    <br><br> 5. Q3 -- 75th percentile
    <br><br> 6. "Outliers"
    <br><br> 7. If data is symmetrical
    <br><br> 8. How tightly data is grouped
    <br><br> 9. If and how your data is skewed

https://builtin.com/data-science/boxplot  -- Good link...


```python
# Box Plot for numerical data  -- Univariate analasis
## orient -- orientation -- "v" -- vertical and "h" -- horizontal

sns.boxplot(df["tip"],orient="h")
```

```python
# Box Plot for numerical data  -- Univariate analasis
## orient -- orientation -- "v" -- vertical and "h" -- horizontal

sns.boxplot(df["total_bill"], orient="v")
```

```python
sns.catplot(x= "day", y = "total_bill", kind = "box", data = df)
```

```python
sns.catplot(x= "day", y = "total_bill", kind = "box", hue = "sex", data = df)
```

```python
# Box Plot for numerical data

sns.boxplot(x = "tip")
```
