<font size  = 5.5> Seaborn is a library for making statistical graphcs in PYthon. It builds on top of matplotlib and integrates closely with pandas data structures... 

```python
import seaborn as sns
```

```python
help(sns.load_dataset)
```

```python
df  = sns.load_dataset("tips")   #### Try using flights dataset...
df
```

```python
### Bar Graph
sns.countplot(data = df, x = "sex")
```

```python
sns.set_theme()
```

```python
### Bar Graph
sns.countplot(data = df, x = "sex")
```

```python
sns.set_theme(style = "whitegrid")
```

```python
sns.barplot(data = df , x = "sex" , y = "tip")
```

```python
sns.barplot(data = df, x = "time" , y ="total_bill")   #### Try giving hue = "sex"
```

```python
sns.barplot(data = df, x = "time" , y ="total_bill", hue = "sex")   #### Try giving hue = "sex"
```

```python
#### Histogram 
sns.histplot(data = df, x = "total_bill")
```

```python
#### Histogram 
sns.histplot(data = df, x = "total_bill", hue = "sex")
```

```python
### Scatter plot
sns.scatterplot(data = df, x = "total_bill", y = "tip")   ### Try giving, hue = "smoker"
```

```python
### Scatter plot
sns.scatterplot(data = df, x = "total_bill", y = "tip", hue = "smoker")   ### Try giving, hue = "smoker"
```

```python
### Line Plot

sns.set_theme()
sns.lineplot(data = df, x = "total_bill", y = "tip")
```

```python
### Line Plot

sns.set_theme()
sns.lineplot(data = df, x = "total_bill", y = "tip", hue = "sex")
```
